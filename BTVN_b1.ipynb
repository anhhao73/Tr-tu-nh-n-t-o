```python
import folium       # Bài 23.2
from geopy.distance import geodesic

# Điểm trung tâm (Ví dụ: Kho trung tâm tại UEH)
center_point = (10.7613, 106.6687)

# Giả lập 10 điểm đến (vĩ độ, kinh độ)
destinations = [
    {"id": "Kho 1", "coords": (10.771, 106.671)},
    {"id": "Kho 2", "coords": (10.752, 106.652)},
    {"id": "Kho 3", "coords": (10.765, 106.680)},
    {"id": "Kho 4", "coords": (10.748, 106.660)},
    {"id": "Kho 5", "coords": (10.780, 106.690)},
    {"id": "Kho 6", "coords": (10.750, 106.685)},
    {"id": "Kho 7", "coords": (10.775, 106.655)},
    {"id": "Kho 8", "coords": (10.785, 106.665)},
    {"id": "Kho 9", "coords": (10.745, 106.675)},
    {"id": "Kho 10", "coords": (10.760, 106.645)}
]

m_23_2 = folium.Map(location=center_point, zoom_start=13)

# Đánh dấu điểm trung tâm
folium.Marker(center_point, popup="Trung tâm", icon=folium.Icon(color='red')).add_to(m_23_2)

# Tính khoảng cách và vẽ đường nối
for dest in destinations:
    dist = geodesic(center_point, dest["coords"]).kilometers

    # Đánh dấu điểm đến với khoảng cách trong popup
    folium.Marker(
        dest["coords"],
        popup=f"{dest['id']}<br>Cách trung tâm: {dist:.2f} km"
    ).add_to(m_23_2)

    # Vẽ PolyLine nối từ trung tâm đến các điểm
    folium.PolyLine(
        locations=[center_point, dest["coords"]],
        color="blue",
        weight=2,
        opacity=0.6,
        tooltip=f"{dist:.2f} km"
    ).add_to(m_23_2)

m_23_2
```

Dựa vào đây, nhà quản trị chuỗi cung ứng có thể đưa ra quyết định mở thêm trạm trung chuyển mini (hub), tăng cường nhân sự giao nhận tại khu vực này để giảm thời gian lấy/giao hàng, đồng thời tối ưu hóa luồng logistics nội đô.

```python
import pandas as pd         # Bài 23.4
import geopandas as gpd
import folium
import numpy as np

# Đọc ranh giới hành chính Việt Nam từ nguồn public (GeoJSON)
url_geojson = "https://raw.githubusercontent.com/TungTh/tungth.github.io/master/data/vn-provinces.json"
gdf = gpd.read_file(url_geojson)

# Lọc ra một số tỉnh/thành phía Nam để bản đồ tập trung hơn
south_provinces = ['Hồ Chí Minh', 'Đồng Nai', 'Bình Dương', 'Long An', 'Bà Rịa - Vũng Tàu', 'Tây Ninh']
gdf_south = gdf[gdf['Ten'].isin(south_provinces)].copy()

# Giả lập số lượng đơn hàng / mức độ tiêu thụ cho các tỉnh này
gdf_south['DonHang'] = np.random.randint(1000, 10000, size=len(gdf_south))

# Tạo bản đồ cơ sở
m_23_4 = folium.Map(location=[10.8, 106.7], zoom_start=8)

# Vẽ Choropleth
folium.Choropleth(
    geo_data=gdf_south,
    name='choropleth',
    data=gdf_south,
    columns=['Ten', 'DonHang'],
    key_on='feature.properties.Ten',
    fill_color='YlOrRd', # Thang màu từ Vàng sang Đỏ
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Số lượng đơn hàng'
).add_to(m_23_4)

folium.LayerControl().add_to(m_23_4)
m_23_4
```

Những tỉnh có màu đậm hơn mang lại doanh thu cao nhất, giúp doanh nghiệp định hướng chiến lược phân bổ lại kho bãi quy mô lớn (Fulfillment Center) gần các vùng trọng điểm này để giảm thiểu chi phí logistics.

```python
import osmnx as ox      # Bài 23.6
import networkx as nx

# Tải dữ liệu mạng lưới đường bộ dành cho ô tô tại một khu vực cụ thể
place_name = "District 10, Ho Chi Minh City, Vietnam"
G = ox.graph_from_place(place_name, network_type='drive')

# Phân tích cơ bản
stats = ox.basic_stats(G)
print("=== PHÂN TÍCH MẠNG LƯỚI GIAO THÔNG KHU VỰC ===")
print(f"Số lượng nút giao (nodes): {stats['n']}")
print(f"Số lượng đoạn đường (edges): {stats['m']}")
print(f"Chiều dài đường trung bình (m): {stats['street_length_avg']:.2f}")

# Trực quan hóa mạng lưới trên bản đồ tĩnh
fig, ax = ox.plot_graph(G, node_size=5, edge_linewidth=0.5, bgcolor='white', node_color='red', edge_color='black')
```

**Vai trò của dữ liệu mạng lưới trong hệ thống AI đô thị thông minh:** Dữ liệu mạng giao thông là "cột sống" của các đô thị thông minh. Các hệ thống AI sử dụng cấu trúc đồ thị (graph) này để mô phỏng luồng phương tiện, dự báo ùn tắc, đồng bộ hóa đèn tín hiệu giao thông tự động và tối ưu hóa tuyến đường cho xe cứu thương/cứu hỏa hoặc xe tự hành trong thời gian thực.

```python
import osmnx as ox        # Bài 23.7
import networkx as nx
import time
import math # Import thêm thư viện math mặc định của Python

# Lấy ngẫu nhiên một điểm bắt đầu và một điểm kết thúc từ đồ thị G (ở câu 23.6)
nodes = list(G.nodes())
orig_node = nodes[0]
dest_node = nodes[-1]

# 1. Thuật toán Dijkstra (Chỉ dùng trọng số khoảng cách)
start_time = time.time()
route_dijkstra = nx.shortest_path(G, orig_node, dest_node, weight='length', method='dijkstra')
time_dijkstra = time.time() - start_time

# 2. Thuật toán A* # SỬA LỖI TẠI ĐÂY: Tự tính khoảng cách Heuristic bằng math.hypot
def heuristic(u, v):
    y1, x1 = G.nodes[u]['y'], G.nodes[u]['x']
    y2, x2 = G.nodes[v]['y'], G.nodes[v]['x']

    # Tính khoảng cách xấp xỉ trên mặt phẳng.
    # Nhân với 111000 để đổi từ đơn vị tọa độ (độ) sang đơn vị mét cho khớp với trọng số đồ thị.
    return math.hypot(x1 - x2, y1 - y2) * 111000

start_time = time.time()
route_astar = nx.astar_path(G, orig_node, dest_node, heuristic=heuristic, weight='length')
time_astar = time.time() - start_time

print(f"Thời gian chạy Dijkstra: {time_dijkstra:.5f} giây")
print(f"Thời gian chạy A*: {time_astar:.5f} giây")
print(f"Độ dài lộ trình (số nút đi qua): {len(route_dijkstra)}")

# Trực quan hóa lộ trình tìm được
fig, ax = ox.plot_graph_route(G, route_dijkstra, route_linewidth=3, node_size=0, bgcolor='white', route_color='blue')
```

Phân tích ưu - nhược điểm:


* **Dijkstra:** Chắc chắn tìm được đường đi ngắn nhất tuyệt đối vì nó duyệt tỏa ra mọi hướng. Nhược điểm là chậm và tốn tài nguyên máy tính vì duyệt qua rất nhiều nút không cần thiết.
* **A (A-Star):**Nhanh hơn Dijkstra đáng kể vì có hàm Heuristic định hướng quá trình tìm kiếm đi thẳng về phía đích. Nhược điểm là nếu hàm Heuristic không chuẩn xác, nó có thể bỏ lỡ đường đi ngắn nhất thực sự. Thuật toán này rất phù hợp cho ứng dụng dẫn đường GPS thực tế.



```python
import folium       # Bài 23.8
import numpy as np
from scipy.optimize import linear_sum_assignment
from geopy.distance import geodesic

# Tọa độ trung tâm giả định
center = [10.771, 106.668]

# Giả lập vị trí 5 khách hàng và 5 xe đang rảnh
np.random.seed(1)
customers = np.column_stack((np.random.normal(10.77, 0.015, 5), np.random.normal(106.66, 0.015, 5)))
drivers = np.column_stack((np.random.normal(10.77, 0.015, 5), np.random.normal(106.66, 0.015, 5)))

# Xây dựng ma trận khoảng cách giữa mọi khách và mọi xe
cost_matrix = np.zeros((5, 5))
for i, c in enumerate(customers):
    for j, d in enumerate(drivers):
        cost_matrix[i, j] = geodesic(c, d).meters

# Giải bài toán gán tuyến tối ưu (Linear Sum Assignment)
row_ind, col_ind = linear_sum_assignment(cost_matrix)

# Trực quan hóa
m_23_8 = folium.Map(location=center, zoom_start=14)

for i in range(5):
    # Marker khách hàng (Icon người màu đỏ)
    folium.Marker(customers[i], popup=f"Khách hàng {i}", icon=folium.Icon(color='red', icon='user')).add_to(m_23_8)
    # Marker xe công nghệ (Icon ô tô màu xanh)
    folium.Marker(drivers[i], popup=f"Tài xế {i}", icon=folium.Icon(color='blue', icon='car')).add_to(m_23_8)

# Vẽ đường nối hiển thị kết quả ghép cuốc
for i, j in zip(row_ind, col_ind):
    folium.PolyLine([customers[i], drivers[j]], color='green', weight=2.5, dash_array='5').add_to(m_23_8)

m_23_8
```

```python
import folium       # Bài 23.9
import numpy as np
from sklearn.cluster import KMeans

# Giả lập vị trí 150 điểm giao hàng phân tán
np.random.seed(42)
orders = np.column_stack((np.random.normal(10.78, 0.03, 150), np.random.normal(106.68, 0.03, 150)))

# Áp dụng K-Means để chia làm 3 khu vực phân phối
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(orders)
centroids = kmeans.cluster_centers_ # Tọa độ tâm cụm

# Trực quan hóa
m_23_9 = folium.Map(location=[10.78, 106.68], zoom_start=12)
colors = ['blue', 'green', 'purple']

# Vẽ các điểm giao hàng theo màu của cụm
for i, order in enumerate(orders):
    folium.CircleMarker(
        location=order,
        radius=4,
        color=colors[labels[i]],
        fill=True,
        fill_opacity=0.6
    ).add_to(m_23_9)

# Vẽ các Hub (Tâm cụm) bằng Marker màu đỏ
for i, centroid in enumerate(centroids):
    folium.Marker(
        location=centroid,
        popup=f"<b>Trạm trung chuyển (Hub) {i+1}</b>",
        icon=folium.Icon(color='red', icon='home')
    ).add_to(m_23_9)

m_23_9
```

**Logic quản trị đi kèm:** Bằng cách gom cụm (clustering), nhà quản trị chuỗi cung ứng có thể định vị chính xác vị trí đặt các trạm trung chuyển (Hub). Tâm của mỗi cụm chính là "trọng tâm" của nhu cầu giao nhận trong khu vực đó. Việc đặt trạm ở tâm cụm sẽ tối thiểu hóa được tổng quãng đường (và chi phí xăng xe, thời gian) di chuyển từ kho đến mọi điểm giao hàng trong cụm. Mô hình này rất phù hợp để quy hoạch mạng lưới kho vệ tinh nội đô (urban micro-fulfillment centers).

```python
# Bài 23.10
# Xây dựng bản đồ nguy cơ tắc nghẽn giao thông và đề xuất tuyến thay thế

!pip install -q folium

import folium
import numpy as np
import pandas as pd

# ----- 1. Tạo dữ liệu giả lập cho 25 điểm giao lộ tại Hà Nội -----
np.random.seed(42)
n_points = 25
df = pd.DataFrame({
    'lat': np.random.uniform(20.99, 21.05, n_points),
    'lon': np.random.uniform(105.80, 105.88, n_points),
    'density': np.random.randint(10, 100, n_points),   # mật độ xe (10-100)
    'hour': np.random.choice([7,8,9,12,15,17,18,19,22], n_points)  # giờ trong ngày
})

# ----- 2. Logic mờ đánh giá nguy cơ tắc nghẽn -----
def fuzzy_risk(density, hour):
    score = 0
    # Mật độ xe
    if density > 70: score += 3
    elif density > 40: score += 2
    else: score += 1
    # Giờ cao điểm
    if (7 <= hour <= 9) or (17 <= hour <= 19): score += 3
    elif 10 <= hour <= 16: score += 1
    else: score += 0
    # Phân loại
    if score >= 5: return 'Cao'
    elif score >= 3: return 'Trung bình'
    else: return 'Thấp'

df['risk'] = df.apply(lambda r: fuzzy_risk(r['density'], r['hour']), axis=1)

# ----- 3. Tạo bản đồ với các điểm nguy cơ -----
m = folium.Map(location=[21.0285, 105.8542], zoom_start=12, tiles='CartoDB positron')
colors = {'Thấp': 'green', 'Trung bình': 'orange', 'Cao': 'red'}

for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row['lat'], row['lon']],
        radius=10,
        color=colors[row['risk']],
        fill=True,
        fillColor=colors[row['risk']],
        fillOpacity=0.7,
        popup=f"Mật độ: {row['density']}<br>Giờ: {row['hour']}h<br>Nguy cơ: {row['risk']}"
    ).add_to(m)

# ----- 4. Đề xuất tuyến đường thay thế (tránh điểm 'Cao') -----
safe = df[df['risk'] != 'Cao'].copy()
if len(safe) >= 2:
    # Sắp xếp các điểm an toàn theo kinh độ để tạo tuyến hợp lý
    route = safe.sort_values('lon')[['lat','lon']].values.tolist()
    folium.PolyLine(route, color='blue', weight=5, opacity=0.8,
                    popup='Tuyến thay thế (tránh điểm tắc nghẽn)').add_to(m)
    folium.Marker(route[0], popup='Bắt đầu', icon=folium.Icon(color='green', icon='play')).add_to(m)
    folium.Marker(route[-1], popup='Kết thúc', icon=folium.Icon(color='red', icon='stop')).add_to(m)

# Hiển thị bản đồ
m
```

```python
import numpy as np          # Bài 23.11
import pandas as pd
import folium
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

# =============================
# 2. TẠO DỮ LIỆU GIẢ LẬP
# =============================
np.random.seed(42)

n = 500

data = pd.DataFrame({
    "lat": np.random.uniform(10.7, 10.9, n),   # khu vực HCM
    "lon": np.random.uniform(106.6, 106.8, n),
    "hour": np.random.randint(0, 24, n),
    "day": np.random.randint(1, 8, n),
})

# nhu cầu giả lập (logic đơn giản)
data["demand"] = (
    50
    + 10 * np.sin(data["hour"] / 24 * 2 * np.pi)  # theo giờ
    + 5 * data["day"]                             # theo ngày
    + np.random.normal(0, 5, n)
)

# =============================
# 3. TRAIN MODEL
# =============================
X = data[["lat", "lon", "hour", "day"]]
y = data["demand"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestRegressor()
model.fit(X_train, y_train)

# =============================
# 4. DỰ ĐOÁN
# =============================
data["predicted_demand"] = model.predict(X)

# =============================
# 5. HIỂN THỊ BẢN ĐỒ
# =============================
map_center = [10.8, 106.7]
m = folium.Map(location=map_center, zoom_start=11)

for _, row in data.sample(200).iterrows():
    color = "green"
    if row["predicted_demand"] > 70:
        color = "red"
    elif row["predicted_demand"] > 60:
        color = "orange"

    folium.CircleMarker(
        location=[row["lat"], row["lon"]],
        radius=5,
        color=color,
        fill=True,
        fill_opacity=0.7,
        popup=f"Nhu cầu: {row['predicted_demand']:.1f}"
    ).add_to(m)

m
```

```python
import folium           # Bài 23.12
import numpy as np
import math

# Dữ liệu 2 Kho hàng (Depots) và 15 Điểm giao nhận (Customers)
depots = [(10.77, 10.66), (10.78, 10.69)]
np.random.seed(10)
customers = [(np.random.uniform(10.75, 10.80), np.random.uniform(10.65, 10.70)) for _ in range(15)]

def distance(p1, p2):
    # Đổi sang mét để tính toán thực tế hơn (1 độ ~ 111,000 mét)
    return math.hypot(p1[0]-p2[0], p1[1]-p2[1]) * 111000

# ==========================================
# PHƯƠNG ÁN CHƯA TỐI ƯU (Baseline)
# 1 xe từ Kho 1 đi giao tuần tự theo danh sách khách hàng ban đầu, sau đó quay về.
# ==========================================
unoptimized_route = [depots[0]] + customers + [depots[0]]
unoptimized_distance = 0
for i in range(len(unoptimized_route)-1):
    unoptimized_distance += distance(unoptimized_route[i], unoptimized_route[i+1])

# ==========================================
# PHƯƠNG ÁN ĐÃ TỐI ƯU (AI / Heuristic)
# ==========================================
# Bước 1: Gán điểm giao hàng cho kho gần nhất (Clustering)
cluster_0, cluster_1 = [], []
for c in customers:
    if distance(c, depots[0]) < distance(c, depots[1]):
        cluster_0.append(c)
    else:
        cluster_1.append(c)

# Bước 2: Heuristic Nearest Neighbor giải bài toán đường đi (TSP) cho từng cụm
def solve_tsp(depot, nodes):
    route = [depot]
    unvisited = nodes.copy()
    current = depot
    while unvisited:
        next_node = min(unvisited, key=lambda x: distance(current, x))
        route.append(next_node)
        unvisited.remove(next_node)
        current = next_node
    route.append(depot)
    return route

route_0 = solve_tsp(depots[0], cluster_0)
route_1 = solve_tsp(depots[1], cluster_1)

# Tính tổng quãng đường phương án tối ưu
def calc_route_distance(route):
    dist = 0
    for i in range(len(route)-1):
        dist += distance(route[i], route[i+1])
    return dist

optimized_distance = calc_route_distance(route_0) + calc_route_distance(route_1)

# ==========================================
# ĐÁNH GIÁ HIỆU QUẢ VÀ IN KẾT QUẢ
# ==========================================
savings_percent = ((unoptimized_distance - optimized_distance) / unoptimized_distance) * 100

print("=== ĐÁNH GIÁ HIỆU QUẢ TỐI ƯU HÓA ===")
print(f"Tổng quãng đường CHƯA tối ưu: {unoptimized_distance/1000:.2f} km")
print(f"Tổng quãng đường ĐÃ tối ưu  : {optimized_distance/1000:.2f} km")
print(f"=> Quãng đường tiết kiệm được : {(unoptimized_distance - optimized_distance)/1000:.2f} km")
print(f"=> Tỷ lệ tối ưu hóa           : Giảm thiểu được {savings_percent:.2f}% tổng quãng đường!\n")

# ==========================================
# TRỰC QUAN HÓA BẢN ĐỒ KẾT QUẢ TỐI ƯU
# ==========================================
m_23_12 = folium.Map(location=[10.775, 10.675], zoom_start=13)

folium.Marker(depots[0], popup="Kho 1", icon=folium.Icon(color='red', icon='home')).add_to(m_23_12)
folium.Marker(depots[1], popup="Kho 2", icon=folium.Icon(color='blue', icon='home')).add_to(m_23_12)

for c in cluster_0: folium.CircleMarker(c, radius=5, color='red', fill=True).add_to(m_23_12)
for c in cluster_1: folium.CircleMarker(c, radius=5, color='blue', fill=True).add_to(m_23_12)

folium.PolyLine(route_0, color='red', weight=2.5, opacity=0.8).add_to(m_23_12)
folium.PolyLine(route_1, color='blue', weight=2.5, opacity=0.8).add_to(m_23_12)

m_23_12
```

```python
import folium         # Bài 23.15
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans

# 1. TẠO DỮ LIỆU ĐƠN HÀNG GIẢ LẬP (Khu vực trung tâm TP.HCM)
np.random.seed(42)
n_orders = 300
# Phân bố tập trung quanh khu vực Quận 10, Quận 1, Quận 3
lats = np.random.normal(10.77, 0.02, n_orders)
lons = np.random.normal(10.68, 0.02, n_orders)
df_orders = pd.DataFrame({'lat': lats, 'lon': lons})

# 2. ÁP DỤNG AI: Thuật toán K-Means Clustering
# Giả sử Giám đốc muốn mở 4 kho vệ tinh (Micro-hubs)
k_hubs = 4
kmeans = KMeans(n_clusters=k_hubs, random_state=42, n_init=10)
df_orders['zone'] = kmeans.fit_predict(df_orders[['lat', 'lon']])
hubs = kmeans.cluster_centers_

# 3. TRỰC QUAN HÓA DASHBOARD BẰNG FOLIUM
# Khởi tạo bản đồ với giao diện tối (Dark Matter) để làm nổi bật dữ liệu
m_void = folium.Map(location=[10.77, 10.68], zoom_start=13, tiles='CartoDB dark_matter')

# Bảng màu đại diện cho 4 khu vực quản lý
colors = ['#FF3366', '#00CC99', '#3399FF', '#FFCC00']

# Vẽ các đơn hàng
for idx, row in df_orders.iterrows():
    folium.CircleMarker(
        location=[row['lat'], row['lon']],
        radius=3,
        color=colors[int(row['zone'])],  # <--- THÊM int() VÀO ĐÂY
        fill=True,
        fill_opacity=0.7,
        weight=0
    ).add_to(m_void)

# Vẽ các Kho vệ tinh (Tâm cụm) bằng Marker lớn
for i, hub in enumerate(hubs):
    html_popup = f"""
    <div style="font-family: Arial; width: 150px;">
        <h4 style="color: {colors[i]}; margin-bottom: 5px;">Kho Vệ Tinh {i+1}</h4>
        <b>Tọa độ:</b> {hub[0]:.4f}, {hub[1]:.4f}<br>
        <b>Số đơn đảm nhận:</b> {len(df_orders[df_orders['zone'] == i])}
    </div>
    """
    folium.Marker(
        location=[hub[0], hub[1]],
        popup=folium.Popup(html_popup, max_width=200),
        icon=folium.Icon(color='white', icon_color=colors[i], icon='building', prefix='fa')
    ).add_to(m_void)

# Tiêu đề Dashboard
title_html = '''
    <h3 align="center" style="font-size:20px; color:white; font-family:Arial; margin-top: 20px;">
        <b>VOID LOGISTICS: MÔ HÌNH PHÂN VÙNG KHO VỆ TINH BẰNG AI (K-MEANS)</b>
    </h3>
'''
m_void.get_root().html.add_child(folium.Element(title_html))

m_void
```
